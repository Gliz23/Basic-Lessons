```python
def f(x,y,f,p):
    A= (np.sin(f*x))**p
    B= (np.sin(f*y))**p
    C= (A+B)/2
    return C
```

```python
x=np.linspace(-np.pi, np.pi)
y=np.linspace(-np.pi, np.pi)

X,Y = np.meshgrid(x,y)
Z= f(X,Y,8,8)
```

```python
fig = plt.figure()
plt.contourf(X, Y,Z, 50, cmap="hot")
```
