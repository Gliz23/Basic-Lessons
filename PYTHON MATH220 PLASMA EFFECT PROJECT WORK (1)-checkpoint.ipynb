# Plasma Effect

Plasma Effect: Plasma effects are often used to create wobbly animations in demo products like screen savers. Using mathematical functions like sine, tan, etc, is one way of creating such effects.

(a) Create a grid with x, y ∈ [−π, π]. Plot the function f(x, y) = sin (5 ∗ x). This an over simplified plasma. 2

We first import the modules and libraries we will need.

```python
import numpy as np
import matplotlib.pyplot as plt
```

We create a grid using the interval -pi and pi by applying linspace to it. We already have two intervals, (x and y) so in order to apply the fuctions we will be given we need a mesh grid. (a1) is a unique variable name which plt.show has been assinged to


```python
#A) We first define the grid with x and y values ranging
#from  -pi to pi and leave the number of values to be defined
# by the program.

x=np.linspace(-np.pi, np.pi)
y=np.linspace(-np.pi, np.pi)

X,Y = np.meshgrid(x,y)

# Define the function f(x,y) = sin(5*x)
Z1 = np.sin(5*X)

# We create the plot in 3d
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(X, Y, Z1)

#We set the labels of the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# We show the 3d plot
a1 = plt.show

#We view the plot in 2d from the z direction
fig = plt.figure()
plt.contourf(X, Y, Z1, 32)
```

C) The next function to plot is f(x,y) = |cos(20(cos^2(20) + sin^2(20))).

(e5) is a unique variable name plt.show for this particular lines of code has been assinged to.

```python
# We define the grid with x and y values ranging from -pi to  pi
x=np.linspace(-np.pi, np.pi)
y=np.linspace(-np.pi, np.pi)

X,Y = np.meshgrid(x,y)

# We define the function f(x,y) = |cos(20(cos^2(20) + sin^2(20)))
def g(x,y):
    W =  (np.cos(20*x)**2) + (np.sin(20*y)**2)
    V = np.cos(20*W)
    return V

# We instantiate the function and assign it to Z5
Z5= g(X,Y)

# We plot the function in 3d
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z5)

# Label the axes 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot in 3d
e5 = plt.show()

# We then show the plot in 2d and view in the z direction
fig = plt.figure()
plt.contourf(X, Y, Z5, 50)
```

Now animate a plasma sequence using the following sequence

v1 = sin (10x + 5t)

v2 = sin (10 (x sin (t/2) + y cos (t/3)) + 5t)

cx = x + *5 sin (t/5)

cy = y + *5 cos (t/3)

v3 = np.sin(((100*(cx^2 + cy^2)+1)^0.5) +t)

v = v1 + v2 + v3

where t is the animation time.
Show the image as a colour image using the following colour component
r = 1
g = cos (πv)
b = sin (πv)

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Define the grid with x and y values ranging from -pi to pi
x = np.linspace(-np.pi, np.pi, 200)
y = np.linspace(-np.pi, np.pi, 200)
X, Y = np.meshgrid(x, y)

# Define the animation time range
t_range = np.linspace(0, 10, 100)

# Set up the plot
fig3 = plt.figure()
 
# Define the animation function
def animate_plasma(t):
    # Define the parameters for the plasma sequence
    v1 = np.sin(10 * X + 5 * t)
    v2 = np.sin(10 * (X * np.sin(t/2) + Y * np.cos(t/3)) + 5 * t)
    cx = X + 0.5 * np.sin(t / 5)
    cy = Y + 0.5 * np.cos(t / 3)
    v3 = np.sin(np.sqrt(100 * (cx**2 + cy**2)) + 1 + t)
    v = v1 + v2 + v3
    r = np.ones_like(v)
    g = np.cos(np.pi * v)
    b = np.sin(np.pi * v)
    img = np.dstack((r,g,b))
    return plt.imshow(img)
    
    
#Create the animation
anim = animation.FuncAnimation(fig3, animate_plasma, frames=200, interval=200, blit = True)

# Show the animation
z = plt.show() 
```
